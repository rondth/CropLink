```python
from pathlib import Path
import pandas as pd
import numpy as np

# If notebook is at testing/wfp_food_prices.ipynb:
# parents[1] => project root
ROOT = Path.cwd().parents[0] if Path.cwd().name == "testing" else Path.cwd()
DATASET_DIR = ROOT / "dataset"

input_csv = DATASET_DIR / "wfp_food_prices_global_2026.csv"

df = pd.read_csv(input_csv)
```
